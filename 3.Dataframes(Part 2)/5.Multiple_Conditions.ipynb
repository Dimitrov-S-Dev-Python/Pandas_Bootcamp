Source: https://www.kaggle.com/datasets/vinicius150987/titanic3
- survival-Survival(0= No;1=Yes)
- class-Passenger Class(1=1st;2=2nd;3=3rd)
- name-Name
- sex-Sex
- age-Age
- sibsp-Number of Siblings/Spouses Aboard
- parch-Number of Parents/Children board
- ticket-Ticket Number
- fare-Passenger Fare
- cabin-Cabin
- embarked-Port of Embarkation(C=Cherbourg;Q=Queenstown;S=Southampton)
- boat-Lifeboat(if survived)
- body-Body number(if not survived and body was recovered)

```python
import pandas as pd
```

```python
cols = ['name', 'survived', 'pclass', 'fare', 'age', 'sex']
```

```python
titanic = pd.read_excel('titanic3.xls', index_col='name', usecols=cols)
titanic
```

Let's filter all females that were in 1st class and also survived

```python
female_mask = titanic['sex'] == 'female'
first_class_mask = titanic['pclass'] == 1
survived_mask = titanic['survived'] == 1
```

```python
titanic[female_mask & first_class_mask & survived_mask]
```

```python
# Select the richest kids on Titanic
rich_kids_mask = (titanic['age'] < 21) & (titanic['fare'] > 200)
rich = titanic[rich_kids_mask]
rich.nlargest(n=5, columns=['age', 'fare'])
```

```python
# Selecting all the people that were under 21 or female that survived
survived_mask = titanic['survived'] == 1
female_mask = titanic['sex'] == 'female'
age_mask = titanic['age'] <= 21
titanic = titanic[(age_mask | female_mask) & survived_mask]
titanic
```

```python
male_mask = titanic['sex'] == 'male'
titanic[male_mask]
```
