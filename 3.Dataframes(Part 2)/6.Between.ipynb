Source: https://www.kaggle.com/datasets/vinicius150987/titanic3
- survival-Survival(0= No;1=Yes)
- class-Passenger Class(1=1st;2=2nd;3=3rd)
- name-Name
- sex-Sex
- age-Age
- sibsp-Number of Siblings/Spouses Aboard
- parch-Number of Parents/Children board
- ticket-Ticket Number
- fare-Passenger Fare
- cabin-Cabin
- embarked-Port of Embarkation(C=Cherbourg;Q=Queenstown;S=Southampton)
- boat-Lifeboat(if survived)
- body-Body number(if not survived and body was recovered)

```python
import pandas as pd
```

```python
cols = ['name', 'survived', 'pclass', 'fare', 'age', 'sex']
```

```python
titanic = pd.read_excel('titanic3.xls', index_col='name', usecols=cols)
titanic
```

```python
upper_age_mask = titanic['age'] >= 20
lower_age_mask = titanic['age'] <= 25
titanic[upper_age_mask & lower_age_mask]
```

```python
youth_mask = titanic['age'].between(20, 25)
titanic[youth_mask]
```

```python
youth_mask = titanic['age'].between(20, 25, inclusive='both')
titanic[youth_mask]
```
