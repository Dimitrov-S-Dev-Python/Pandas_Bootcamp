Source: https://www.kaggle.com/datasets/vinicius150987/titanic3
- survival-Survival(0= No;1=Yes)
- class-Passenger Class(1=1st;2=2nd;3=3rd)
- name-Name
- sex-Sex
- age-Age
- sibsp-Number of Siblings/Spouses Aboard
- parch-Number of Parents/Children board
- ticket-Ticket Number
- fare-Passenger Fare
- cabin-Cabin
- embarked-Port of Embarkation(C=Cherbourg;Q=Queenstown;S=Southampton)
- boat-Lifeboat(if survived)
- body-Body number(if not survived and body was recovered)

```python
import pandas as pd
```

```python
titanic = pd.read_excel('titanic3.xls', index_col='name')
titanic
```

```python
titanic.sort_index(inplace=True)
titanic
```

```python
titanic.loc['Abbott, Mr. Rossmore Edward'] # Selecting a row via index
```

```python
titanic.loc[['Abbott, Mr. Rossmore Edward', 'Zabour, Miss. Hileni']] 
# Selecting multiple rows via index
```

```python
titanic.loc['Federice'] # Will raise an exception because it doesn't exist
```

```python
titanic.loc['Kelly, Mr. James'] # If there are duplicates , it will return them all
```

```python
titanic.loc['Alen, Miss. Elisabeth Walton': 'Allison, Miss. Helen Loraine']
# We can select multiple rows with slicing
```

```python
titanic.loc['Allison, Miss. Helen Loraine', 'fare']
# Returns a value for the passenger for the specific column
```

```python
titanic.loc['Allison, Miss. Helen Loraine', ['fare', 'sex', 'cabin']]
# Retirn a series of the passenger we selected with the specific columns
```

```python
passengers = ['Allen, Miss. Elisabeth Walton', 'Allison, Miss. Helen Loraine']
titanic.loc[passengers, 'fare']
```

```python
passengers = ['Allen, Miss. Elisabeth Walton', 'Allison, Miss. Helen Loraine']
titanic.loc[passengers, ['fare', 'pclass', 'survived']]
```
